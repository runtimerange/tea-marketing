```python
df = df.dropna()
df.shape
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df.info()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mpl_dates
```

```python
plt.figure(figsize = (18,6))
plt.plot(df['Date'], df['Close'], color = 'red')
plt.title('Tata-tea stock price(2015 - 2022)', fontsize = 14, fontweight = 'bold')
plt.xlabel("Year", fontsize =13)
plt.ylabel("Price in INR", fontsize = 13)
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (18,6))
plt.bar(df['Date'], df['Close'])
plt.title('Tata tea(2015 - 2022)', fontsize = 14, fontweight = 'bold')
plt.xlabel("Year", fontsize =13)
plt.ylabel("Price in INR", fontsize = 13)
plt.grid()
plt.show()
```

```python
df['Market cap'] = df["Open"]*df["Volume"]
```

```python
df['Volume'] = df['Volume'].astype(int)
df['Market cap'] = df['Market cap'].astype(int)
```

```python
df.head()
```

```python
plt.figure(figsize = (20,6))
plt.plot(df['Date'], df['Market cap'], color ='orange')
plt.title("Market capitalization of Tata-tea(2015-2022)", fontsize = 14, weight = 'bold')
plt.xlabel("Date")
plt.ylabel("Market cap")
plt.grid()
plt.show()
```
